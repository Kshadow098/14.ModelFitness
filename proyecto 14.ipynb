Observaciones clave de la correlaci√≥n:

```python
X = df.drop('Churn',axis = 1)
y = df['Churn']

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.20,random_state=0)
model_regresion = LogisticRegression(solver='liblinear', random_state=0)
model_regresion.fit(X_train,y_train)
y_pred_regresion = model_regresion.predict(X_test)

model_rf = RandomForestClassifier(n_estimators=100, random_state=0)
model_rf.fit(X_train, y_train)
y_pred_rf = model_rf.predict(X_test)

def print_metrics(y_true, y_pred, title):
    print(title)
    print(f"\tAccuracy:  {accuracy_score(y_true, y_pred):.2f}")
    print(f"\tPrecision: {precision_score(y_true, y_pred):.2f}")
    print(f"\tRecall:    {recall_score(y_true, y_pred):.2f}\n")

print_metrics(y_test, y_pred_regresion, "--- Regresi√≥n Log√≠stica ---")
print_metrics(y_test, y_pred_rf, "--- Bosque Aleatorio ---")
```

Utilizaremos el modelo de Regresi√≥n Log√≠stica, porque supero al Bosque Aleatorio en las tres m√©tricas clave (Accuracy, Precision y Recall).Interpretaci√≥n:Tiene un Recall de 0.83, lo que significa que detecta correctamente al 83% de los usuarios que realmente se van a ir. Esto es crucial para la retenci√≥n, ya que nos interesa "pescar" a la mayor cantidad de clientes en riesgo posible.Su Precisi√≥n de 0.86 indica que cuando predice que alguien se ir√°, acierta el 86% de las veces, minimizando el gasto en incentivos para clientes que no pensaban irse (falsos positivos).

```python
scaler = StandardScaler()
X_sc = scaler.fit_transform(X)

linked = linkage(X_sc, method='ward')
plt.figure(figsize=(15, 10))
dendrogram(linked, orientation='top')
plt.title('Dendrograma Jer√°rquico de Clustering')
plt.xlabel('Muestras')
plt.ylabel('Distancia')
plt.show()

km = KMeans(n_clusters=5, random_state=0)
labels = km.fit_predict(X_sc)

df['cluster_km'] = labels

cluster_means = df.groupby('cluster_km').mean()
print("--- Medias de caracter√≠sticas por Cluster ---")
print(cluster_means)

churn_rate = df.groupby('cluster_km')['Churn'].mean()
print("\n--- Tasa de Cancelaci√≥n por Cluster ---")
print(churn_rate)

binary_cols = ['Near_Location', 'Partner', 'Promo_friends', 'Contract_period']
continuous_cols = ['Lifetime', 'Age', 'Avg_additional_charges_total', 'Avg_class_frequency_current_month']

for col in binary_cols:
    plt.figure(figsize=(10, 5))
    sns.countplot(x=col, hue='cluster_km', data=df, palette='viridis')
    plt.title(f'Distribuci√≥n de {col} por Cluster')
    plt.legend(title='Cluster')
    plt.show()

for col in continuous_cols:
    plt.figure(figsize=(10, 5))
    sns.boxplot(x='cluster_km', y=col, data=df, palette='viridis')
    plt.title(f'Distribuci√≥n de {col} por Cluster')
    plt.show()
```
